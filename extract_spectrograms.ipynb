## Load audio file with librosa



```python
audio_gun, __ = librosa.load(audio_gun)
audio_bg, _ = librosa.load(audio_bg)
```

## Exatracting Short-Time FOurier  Transform

```python
gun = librosa.stft(audio_gun)
bg = librosa.stft(audio_bg)
```

```python
type(gun[0][0])
```

## Calculate the spectrogram

```python
y_gun = np.abs(gun) ** 2
y_bg = np.abs(bg) ** 2
```

```python
y_gun
```

```python
y_gun.shape
```

## Visualizing the spectrogram

```python

def plot_spectrogram(y,y_axis="linear"):
    plt.Figure(figsize=(25,10))
    librosa.display.specshow(y,
                             x_axis="time",
                             y_axis=y_axis,)
    plt.colorbar(format="%+2.f")
```

```python
plot_spectrogram(y_gun,)
```

## Log-Amplitude Spectrogram

```python
y_log_gun = librosa.power_to_db(y_gun)
plot_spectrogram(y_log_gun)
```

```python
y_log_bg = librosa.power_to_db(y_bg)
plot_spectrogram(y_log_bg)
```

## LOg frequency Spectrogram

```python
# FOr gun shot
y_log_gun = librosa.power_to_db(y_gun)
plot_spectrogram(y_log_gun,y_axis="log")
```

```python
# FOr backgroung
y_log_bg = librosa.power_to_db(y_bg)
plot_spectrogram(y_log_bg,y_axis="log")
```
