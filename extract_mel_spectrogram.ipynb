```python
import librosa
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
audio_bg = "audio_data/background.WAV"
audio_gun = "audio_data/gun.WAV"
```

```python
ipd.Audio(audio_gun)
```

```python
audio_gun, __ = librosa.load(audio_gun)
```

## Mel Fillter banks

```python
fliter_banks = librosa.filters.mel(n_fft=2048,sr=2205,n_mels=90                                  )
```

```python
fliter_banks.shape
```

```python
def plot_spectrogram(y,y_axis="linear"):
    plt.Figure(figsize=(25,10))
    librosa.display.specshow(y,
                             x_axis="time",
                             y_axis=y_axis,)
    plt.colorbar(format="%+2.f")
```

```python
plot_spectrogram(fliter_banks)
```

## EXtracting mel spectrogram

```python
mel_spectrogram = librosa.feature.melspectrogram(y=audio_gun,n_mels=9)
```

```python
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
log_mel_spectrogram.shape
```

```python
plot_spectrogram(log_mel_spectrogram,y_axis="mel")
```
